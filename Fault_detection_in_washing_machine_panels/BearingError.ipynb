### Feature Extraction/Creation

```python
# Features extracted for each sensor:
# 1.Max 2.Min 3.Mean 4.Standard Deviation 5.Root Mean Square

def extractFeature(df):
    rms_sensor1 = math.sqrt((df['Sensor1'] ** 2).mean())
    rms_sensor2 = math.sqrt((df['Sensor2'] ** 2).mean())
    feature_array = np.array([ df['Sensor1'].max(), df['Sensor2'].max(),df['Sensor1'].min(),df['Sensor2'].min(),df['Sensor1'].mean(), df['Sensor2'].mean(),df['Sensor1'].std(),df['Sensor2'].std(), rms_sensor1, rms_sensor2])
    return feature_array
```

### Reading vibration data file and creating feature matrix

```python
feature_df = pd.DataFrame([])

for file in os.listdir(vibration_folder):
    vib_data = pd.read_csv(vibration_folder/file, sep = "\t", header = None)
    vib_data.columns = ['Sensor1','Sensor2']

    #Function call to extract features
    feature_array = extractFeature(vib_data)
    vib_df = pd.DataFrame([feature_array])
    
    #Converting string filename to datetime object
    file_datetime = datetime.strptime(str(file), '%Y-%m-%d_%H-%M-%S')
    vib_df.index = [file_datetime]
    
    feature_df = feature_df.append(vib_df)
```

```python
feature_df.columns = ['S1_max', 'S2_max','S1_min','S2_min','S1_mean','S2_mean','S1_std','S2_std','S1_rms','S2_rms']
```

### Modelling and Evaluation

```python
features_train = ['S1_max','S1_min','S1_mean','S1_std','S1_rms','S2_max','S2_min','S2_mean','S2_std','S2_rms']
X_features = df_merged[features_train].to_numpy()
Y = df_merged['quality'].to_numpy()

# Encode labels: "OK" - 1 "nOK" - 0

Y = np.where(Y == 'OK',1,0)
```

```python
# Split the feature matrix intro test and train

X_train, X_test, Y_train, Y_test = train_test_split(X_features,Y,test_size = 0.30)
```

```python
# Instantiate the Scaler and the Classifier

scaler = StandardScaler()
logreg = LogisticRegression()
logreg2 = LogisticRegression( solver = 'liblinear', class_weight = 'balanced')
```

```python
# IGNORE: Junk code
# X_train = scaler.fit_transform(X_train)
# X_test = scaler.transform(X_test)
# model.fit(X_train, Y_train)
# preds = model.predict(X_test)
```

```python
#Instantiate a pipeline object with Scaler and Classifier 

pipe_log = Pipeline([('scaler',scaler),("classifier",logreg)])
pipe_svc = Pipeline([('scaler',scaler),("classifier",SVC(class_weight = 'balanced'))])
pipe_log2 = Pipeline([('scaler',scaler),("classifier",logreg2)])
```

```python
# Fit the pipeline object

pipe_log.fit(X_train,Y_train)
pipe_svc.fit(X_train,Y_train)
pipe_log2.fit(X_train,Y_train)
```

```python
preds_log = pipe_log.predict(X_test)
preds_svc = pipe_svc.predict(X_test)
preds_log2 = pipe_log2.predict(X_test)
```

### Evaluation and Metrics

#### Weighted F1

```python
# Should we use weighted average in metrics? It helps account for label imbalance

weighted_f1 = []
weighted_f1.append(metrics.f1_score(Y_test, preds_log, average = 'weighted'))
weighted_f1.append(metrics.f1_score(Y_test, preds_svc,average = 'weighted'))
weighted_f1.append(metrics.f1_score(Y_test, preds_log2,average = 'weighted'))
print(weighted_f1)
```

#### F1 Score

```python
print("Logistic:")
print(metrics.f1_score(Y_test, preds_log, pos_label = 0))
print("SVC Weight")
print(metrics.f1_score(Y_test, preds_svc,pos_label = 0))
print("Logistic Weight:")
print(metrics.f1_score(Y_test, preds_log2,pos_label = 0))
```

#### Recall

```python
recalls = []
recalls.append(metrics.recall_score(Y_test, preds_log, pos_label = 0))
recalls.append(metrics.recall_score(Y_test, preds_svc,pos_label = 0))
recalls.append(metrics.recall_score(Y_test, preds_log2,pos_label = 0))
print(recalls)
```

#### Precision

```python
precisions = []
precisions.append(metrics.precision_score(Y_test, preds_log, pos_label = 0))
precisions.append(metrics.precision_score(Y_test, preds_svc,pos_label = 0))
precisions.append(metrics.precision_score(Y_test, preds_log2,pos_label = 0))
print(precisions)
```

#### Confusion Matrix

```python
#Confusion Matrix for three classifiers

cf_matrix_log = confusion_matrix(Y_test,preds_log)
cf_matrix_svc = confusion_matrix(Y_test,preds_svc)
cf_matrix_log2 = confusion_matrix(Y_test,preds_log2)
```

```python
# Plotting Confusion Matrix

ax = sns.heatmap(cf_matrix_log, annot=True, cmap='Blues', fmt = 'd')
ax.set_title('Confusion Matrix: Log Regressor without Weighting');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
ax.xaxis.set_ticklabels(['0: NotOk','1: OK'])
ax.yaxis.set_ticklabels(['0: NotOk','1: OK'])
plt.show()
```

```python
ax = sns.heatmap(cf_matrix_svc, annot=True, cmap='Blues', fmt = 'd')
ax.set_title('Confusion Matrix: SVC with Weighting');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
ax.xaxis.set_ticklabels(['0: NotOk','1: OK'])
ax.yaxis.set_ticklabels(['0: NotOk','1: OK'])
plt.show()
```

```python
ax = sns.heatmap(cf_matrix_log2, annot=True, cmap='Blues', fmt = 'd')
ax.set_title('Confusion Matrix: LogRegressor with Weighting');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
ax.xaxis.set_ticklabels(['0: NotOk','1: OK'])
ax.yaxis.set_ticklabels(['0: NotOk','1: OK'])
plt.show()
```
